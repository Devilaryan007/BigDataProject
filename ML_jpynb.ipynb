```python
import pandas as pd
from bs4 import BeautifulSoup
import requests

book_titles = []
book_prices = []
book_instock_status = []

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}

base_url = "https://books.toscrape.com/catalogue/page-{}.html"

# Loop through all 50 pages
for page_num in range(1, 51):
  page_url = base_url.format(page_num)
  print(f"Scraping page: {page_url}") # Added print to show progress
  try:
    books_response_text = requests.get(page_url, headers=headers, timeout=30).text
    soup = BeautifulSoup(books_response_text, 'lxml')
    product_pods = soup.find_all("article", class_="product_pod")

    for b in product_pods:
      title = b.find('h3').find('a')['title']
      book_titles.append(title)

      price_tag = b.find('p', class_='price_color')
      price_text = price_tag.text.replace('£', '').replace('Â', '')
      book_prices.append(float(price_text))

      instock_tag = b.find('p', class_='instock availability')
      instock_text = instock_tag.text.strip()
      book_instock_status.append(instock_text)
  except requests.exceptions.RequestException as e:
    print(f"Error scraping {page_url}: {e}")
    continue # Continue to the next page if there's an error

books_df = pd.DataFrame({
    'Title': book_titles,
    'Price': book_prices,
    'Availability': book_instock_status
})
display(books_df.head())
print(f"Total books scraped: {len(books_df)}")
books_df.to_csv('books_data.csv', index=False)
```
